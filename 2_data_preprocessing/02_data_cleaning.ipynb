```python
num_columns = df.select_dtypes(include=['int64', 'float64']).columns
cat_columns = df.select_dtypes(include=['object']).columns
```

```python
for col in num_columns:
    df[col] = df[col].fillna(df[col].median())

for col in cat_columns:
    df[col] = df[col].fillna(df[col].mode()[0])

df.sample(50)
```

```python
df.drop_duplicates(inplace=True)
df.sample(50)
```

```python
"""
for col in cat_columns:

    # convert values to string ........
    df[col] = df[col].astype(str)

    # remove extra spaces from start and end .........
    df[col] = df[col].str.strip()

    # convert to lowercase ........
    df[col] = df[col].str.lower()
"""

for col in cat_columns:
    df[col] = df[col].astype(str).str.strip().str.lower()

df.sample(20)
```

```python
df.to_csv('../datasets/cleaned_data.csv', index=False)
```

steps to Find Outliers Using IQR [interquartile range]:

Find Q1 and Q3

→ Use data.quantile(0.25) and data.quantile(0.75)

Calculate IQR

→ IQR = Q3 - Q1

Set boundaries:

Lower Bound = Q1 – 1.5 × IQR

Upper Bound = Q3 + 1.5 × IQR

Any value below the lower bound OR above the upper bound is considered an OUTLIER.

 The 1.5 multiplier is a standard rule (from statistician John Tukey). You can use 3.0 for "extreme outliers", but 1.5 is common.
