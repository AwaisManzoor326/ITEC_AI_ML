```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

# Set the style of the plots
sns.set_style("darkgrid")
```

```python
df = pd.read_csv('../datasets\coaster_db.csv')
# Jab bhi DataFrame print karun, saari columns full dikhani hain. Kuch bhi hide ya '...' wala short form mat karna.
pd.set_option('display.max_columns', None)
```

```python
num_columns = df.select_dtypes(include=['int64', 'float64']).columns
cat_columns = df.select_dtypes(include=['object']).columns
```

```python
for col in num_columns:
    df[col] = df[col].fillna(df[col].median())

for col in cat_columns:
    df[col] = df[col].fillna(df[col].mode()[0])

df.sample(50)
```

```python
df.drop_duplicates(inplace=True)
df.sample(50)
```

```python
"""
for col in cat_columns:

    # convert values to string ........
    df[col] = df[col].astype(str)

    # remove extra spaces from start and end .........
    df[col] = df[col].str.strip()

    # convert to lowercase ........
    df[col] = df[col].str.lower()
"""

for col in cat_columns:
    df[col] = df[col].astype(str).str.strip().str.lower()

df.sample(20)
```

```python
df.to_csv('../datasets/cleaned_data.csv', index=False)
```
