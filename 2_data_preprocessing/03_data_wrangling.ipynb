```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets\cleaned_data.csv')
```

```python
pd.set_option('display.max_columns', None)
```

```python
df.head(1)
```

```python
df.rename(columns={
    'coaster_name' : 'name',
    'height_ft' : 'height',
    'speed_mph' : 'speed'
}, inplace=True, errors='ignore')

df.head(1)
```

```python
df = df[df['height'] > 50]

df.columns
```

```python
df['speed_kmh'] = df['speed'] * 1.60934

df.columns
```

```python
df.groupby('Manufacturer')['speed'].agg(['mean', 'median', 'max', 'min'])
```

```python
for manuf, speed in df.groupby('Manufacturer'):
    print(manuf)
    print(speed)
```

```python
df.sort_values(by='speed', ascending=False, inplace=True)
```

```python
df.to_csv('../datasets/wrangled_data.csv', index=False)
df.head(2)
```
