```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/wrangled_data.csv')
pd.set_option('display.max_columns', None)

df.head(1)
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

**label  encoding**

```python
MyLabelEncoder = LabelEncoder()

df['manufacturer_encoded'] = MyLabelEncoder.fit_transform(df['Manufacturer'])

df.sample(3)
```

**one hot encoding**

```python
if 'Type' in df.columns:
    df = pd.get_dummies(df, columns=['Type'], drop_first=True)
    
# df.sample(3)
df.columns
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
myScaler = StandardScaler()
scale_col = ['speed', 'height']

for col in scale_col:
    if col in df.columns:
        df[col + '_scaled'] = myScaler.fit_transform(df[[col]])
        
df.sample(3)
```

```python
list(df.columns)
```

```python
df['height_speed_ratio'] = df['height'] / df['speed']
df.head()
```

```python
df.to_csv('../datasets/featured_data.csv', index=False)
df.head(1)
```
