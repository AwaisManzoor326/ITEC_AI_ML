```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('../datasets/moons_dataset_noisy.csv')
df.head(10)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='X1', y='X2', color='blue', s=50)
plt.title('Original Noisy Moons Dataset')
plt.show()
```

```python
X = df[['X1', 'X2']].values
scaled_X = StandardScaler().fit_transform(X)
```

```python
model = DBSCAN(eps=0.3, min_samples=5)
model.fit(scaled_X)
df['Cluster'] = model.labels_
```

```python
plt.figure(figsize=(8,6))
palette = sns.color_palette('Set1', n_colors=len(set(df['Cluster'])))
sns.scatterplot(data=df, x='X1', y='X2', hue='Cluster', palette=palette)
plt.title('DBSCAN Clustering on Noisy Moons Dataset')
plt.show()
```
