```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('../datasets/AgeIncomeEnhanced.csv')
df.head(10)
```

```python
plt.scatter(df['Age'], df['Income'], s=50)
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Age vs Income')
plt.show()
```

```python
features = df[["Age","Income","StudyHours","Spending","WorkHours"]]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(scaled_features)
df['Clusters'] = kmeans.labels_

df.sample(10)
```

```python
plt.scatter(df["Age"], df["Income"], c=df["Clusters"], cmap = "viridis", s = 50)
plt.xlabel("Age")
plt.ylabel("Income")
plt.title("K-Means Clusters")
plt.show()
```
